````python
import csv
import re

# reading recweecweekdata.csv files
with open("recweekdata.csv") as dirty_data:
    dirty_data_reader = csv.reader(dirty_data, delimiter = ",")
    
    # creating empty lists to classify initial data provided
    # data that no longer require cleansing are labelled as "cleanlist"
    cleanlist_firstname = []
    cleanlist_lastname = []
    dirtylist_birthmonth = []
    cleanlist_birthday = []
    dirtylist_birthyear = []
    dirtylist_course = []
    dirtylist_phonenumber = []
    
    # row number accumulator
    row_number = 0
    
    for row in dirty_data_reader:
        # disregard header
        if row_number == 0:
            pass
        # categorizes dirty data into empty lists 
        else:
            cleanlist_firstname.append(row[0])
            cleanlist_lastname.append(row[1])
            dirtylist_birthmonth.append(row[2])
            cleanlist_birthday.append(row[3])
            dirtylist_birthyear.append(row[4])
            dirtylist_course.append(row[5])
            dirtylist_phonenumber.append(row[6])
        row_number += 1
        
        
    # cleans birthmonth dirty dataset
    # assign an official name for each unique name
    unique_birthmonth_dictionary = {"January":"January", "February":"February", "March":"March", "April":"April", "May":"May", "June":"June",
                                    "July":"July", "August":"August", "September":"September", "October":"October", "November":"November",
                                    "December":"December", "Jan":"January", "Feb":"February", "Mar":"March", "Apr":"April", "Jun":"June",
                                    "Jul":"July", "Aug":"August", "Sep":"September", "Oct":"October", "Nov":"November", "Dec":"December",
                                    "1":"January", "2":"February", "3":"March", "4":"April", "5":"May", "6":"June", "7":"July", "8":"August",
                                    "9":"September", "10":"October", "11":"November", "12":"December"}
        
    cleanlist_birthmonth = [re.sub(r''.join(birthmonth),r''.join(unique_birthmonth_dictionary[birthmonth]),birthmonth) for birthmonth in dirtylist_birthmonth]
    
    
    # cleans birthyear dirty dataset
    # assign an official name for each unique name
    unique_birthyear_dictionary = {"1997":"1997", "1998":"1998", "1999":"1999",
                                   "2000":"2000", "2001":"2001", "97":"1997", 
                                   "98":"1998", "99":"1999", "00":"2000", "01":"2001"}
    
    cleanlist_birthyear = [re.sub(r''.join(birthyear),r''.join(unique_birthyear_dictionary[birthyear]),birthyear) for birthyear in dirtylist_birthyear]
    
    
    # cleans course dirty dataset
    # assign an official name for each unique name
    
    unique_course_dictionary = {'ME': 'BS Management Engineering','BS LM': 'BS Legal Management','MGT': 'BS Management',
                         'Communications Technology': 'BS Communications Technology Management','BS MGT': 'BS Management',
                         'Information Technology Entrepreneurship': 'BS Information Technology Entrepreneurship',
                         'BS CTM': 'BS Communications Technology Management','ITE': 'BS Information Technology Entrepreneurship',
                         'BS ITE': 'BS Information Technology Entrepreneurship','Management': 'BS Management',
                         'CTM': 'BS Communications Technology Management','LM': 'BS Legal Management','Legal Management': 'BS Legal Management',
                         'BS ME': 'BS Management Engineering','Management Engineering': 'BS Management Engineering'}
    
    cleanlist_course = [re.sub(r''.join(course),r''.join(unique_course_dictionary[course]),course) for course in dirtylist_course]
    
    
    # cleans phonenumber dirty dataset
    pre_clean_phonenumber = [re.sub(r'^(9[0-9][0-9])',r'+63 \1',phonenumber) for phonenumber in dirtylist_phonenumber]
    cleanlist_phonenumber = []
    for phonenumber in pre_clean_phonenumber:
        cleanlist_phonenumber.append(phonenumber.replace(" ",""))    
    
    
    # combining cleaned lists
    clean_list = list(zip(cleanlist_firstname,cleanlist_lastname,cleanlist_birthmonth,cleanlist_birthday,cleanlist_birthyear,cleanlist_course,cleanlist_phonenumber))
    
    
# cleaned recweekdata.csv called cleanrecweekdata.csv
with open('cleanrecweekdata.csv','w') as clean_data:
    clean_data_writer = csv.writer(clean_data,lineterminator='\n')
    # labels with headers
    clean_data.write('firstname,lastname,birthmonth,birthday,birthyear,course,phonenumber\n')
    # inputs cleansed data
    clean_data_writer.writerows(clean_list)
````
